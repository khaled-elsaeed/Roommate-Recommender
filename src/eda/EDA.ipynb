```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Ensure Age column is ordered correctly
df['Age'] = pd.Categorical(df['Age'], categories=sorted(df['Age'].unique()), ordered=True)

# Define a custom color palette
custom_palette = sns.color_palette("Set2", 4)  # Adjusted to match the number of categories

# Function to annotate bars with percentage
def annotate_percentage(ax, orientation='vertical'):
    total = len(df)
    for p in ax.patches:
        if orientation == 'vertical':
            percentage = f'{100 * p.get_height() / total:.1f}%'
            x = p.get_x() + p.get_width() / 2
            y = p.get_height()
            ax.annotate(percentage, (x, y), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
        elif orientation == 'horizontal':
            percentage = f'{100 * p.get_width() / total:.1f}%'
            x = p.get_width()
            y = p.get_y() + p.get_height() / 2
            ax.annotate(percentage, (x, y), ha='center', va='center', xytext=(20, 0), textcoords='offset points')

# Visualizations with updated column names and attributes

# Gender Distribution
plt.figure(figsize=(8, 4))
ax = sns.countplot(x='Gender', data=df, palette=custom_palette)
plt.title('Gender Distribution', pad=20)  # Increase space between title and plot
annotate_percentage(ax)
ax.spines['top'].set_visible(False)  # Remove top border
ax.spines['right'].set_visible(False)  # Remove right border
plt.tight_layout()
plt.savefig('figures/gender_distribution.png')  # Save figure
plt.show()

# Age Distribution
plt.figure(figsize=(8, 4))
ax = sns.countplot(x='Age', data=df, order=sorted(df['Age'].unique()), palette=custom_palette)
plt.title('Age Distribution', pad=20)  # Increase space between title and plot
annotate_percentage(ax)
ax.spines['top'].set_visible(False)  # Remove top border
ax.spines['right'].set_visible(False)  # Remove right border
plt.tight_layout()
plt.savefig('figures/age_distribution.png')  # Save figure
plt.show()

# Faculty Distribution
plt.figure(figsize=(12, 6))
ax = sns.countplot(y='Faculty', data=df, order=df['Faculty'].value_counts().index, palette=custom_palette)
plt.title('Faculty Distribution', pad=20)  # Increase space between title and plot
annotate_percentage(ax, orientation='horizontal')  # Horizontal annotation for y-axis plot
ax.spines['top'].set_visible(False)  # Remove top border
ax.spines['right'].set_visible(False)  # Remove right border
plt.tight_layout()
plt.savefig('figures/faculty_distribution.png')  # Save figure
plt.show()

# Smoking Habits
plt.figure(figsize=(8, 4))
ax = sns.countplot(x='Smoking Habit', data=df, palette=custom_palette)
plt.title('Smoking Habits', pad=20)  # Increase space between title and plot
annotate_percentage(ax)
ax.spines['top'].set_visible(False)  # Remove top border
ax.spines['right'].set_visible(False)  # Remove right border
plt.tight_layout()
plt.savefig('figures/smoking_habits.png')  # Save figure
plt.show()

# Attitude towards Roommate Smoking
plt.figure(figsize=(8, 4))
ax = sns.countplot(x='Attitude to Roommate Smoking', data=df, palette=custom_palette)
plt.title('Attitude towards Roommate Smoking', pad=20)  # Increase space between title and plot
annotate_percentage(ax)
ax.spines['top'].set_visible(False)  # Remove top border
ax.spines['right'].set_visible(False)  # Remove right border
plt.tight_layout()
plt.savefig('figures/attitude_roommate_smoking.png')  # Save figure
plt.show()

# Preferred Roommate Relationship Type
plt.figure(figsize=(8, 4))
ax = sns.countplot(x='Roommate Relationship Preference', data=df, palette=custom_palette)
plt.title('Preferred Roommate Relationship Type', pad=20)  # Increase space between title and plot
plt.xticks(rotation=45, ha='right')
annotate_percentage(ax)
ax.spines['top'].set_visible(False)  # Remove top border
ax.spines['right'].set_visible(False)  # Remove right border
plt.tight_layout()
plt.savefig('figures/preferred_relationship.png')  # Save figure
plt.show()

# Desired Room Attributes
plt.figure(figsize=(8, 4))
ax = sns.countplot(x='Desired Room Attributes', data=df, palette=custom_palette)
plt.title('Desired Room Attributes', pad=20)  # Increase space between title and plot
plt.xticks(rotation=45, ha='right')
annotate_percentage(ax)
ax.spines['top'].set_visible(False)  # Remove top border
ax.spines['right'].set_visible(False)  # Remove right border
plt.tight_layout()
plt.savefig('figures/desired_attributes.png')  # Save figure
plt.show()

# Conflict Handling Method
plt.figure(figsize=(10, 6))
ax = sns.countplot(y='Conflict Handling Style', data=df, order=df['Conflict Handling Style'].value_counts().index, palette=custom_palette)
plt.title('Conflict Handling Method', pad=20)  # Increase space between title and plot
annotate_percentage(ax, orientation='horizontal')  # Horizontal annotation for y-axis plot
ax.spines['top'].set_visible(False)  # Remove top border
ax.spines['right'].set_visible(False)  # Remove right border
plt.tight_layout()
plt.savefig('figures/conflict_handling.png')  # Save figure
plt.show()

# Roommate Concerns
concern_columns = ['Concerns: Obnoxious Behavior', 'Concerns: Loud Music/TV', 'Concerns: Noise While Studying/Sleeping',
                   'Concerns: Food Consumption', 'Concerns: Cleanliness', 'Concerns: Unfriendliness',
                   'Concerns: Privacy & Respect', 'Concerns: Open-Mindedness', 'Concerns: Personal Hygiene',
                   'Concerns: Item Sharing', 'Concerns: Guests Frequency', 'Concerns: Messiness']

plt.figure(figsize=(10, 10))
concern_data = df[concern_columns].melt(var_name='Concern', value_name='Response')
ax = sns.countplot(y='Concern', hue='Response', data=concern_data, order=concern_data['Concern'].unique(), palette=custom_palette)
plt.title('Roommate Concerns', pad=20)  # Increase space between title and plot
plt.legend(title='Response', loc='upper right')
annotate_percentage(ax, orientation='horizontal')  # Horizontal annotation for y-axis plot
ax.spines['top'].set_visible(False)  # Remove top border
ax.spines['right'].set_visible(False)  # Remove right border
plt.tight_layout()
plt.savefig('figures/roommate_concerns.png')  # Save figure
plt.show()
```
