```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure Age column is ordered correctly
df['Age'] = pd.Categorical(df['Age'], categories=sorted(df['Age'].unique()), ordered=True)

# Define a custom color palette
custom_palette = sns.color_palette("Set2", 4)  # Adjusted to match the number of categories

# Function to annotate bars with percentage
def annotate_percentage(ax, orientation='vertical'):
    total = len(df)
    for p in ax.patches:
        if orientation == 'vertical':
            percentage = f'{100 * p.get_height() / total:.1f}%'
            x = p.get_x() + p.get_width() / 2
            y = p.get_height()
            ax.annotate(percentage, (x, y), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
        elif orientation == 'horizontal':
            percentage = f'{100 * p.get_width() / total:.1f}%'
            x = p.get_width()
            y = p.get_y() + p.get_height() / 2
            ax.annotate(percentage, (x, y), ha='center', va='center', xytext=(20, 0), textcoords='offset points')

# Visualizations for each categorical column

# Personal Room Description
plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='Personal Room Description', data=df, palette=custom_palette)
plt.title('Personal Room Description', pad=20)
annotate_percentage(ax1)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('figures/personal_room_description.png')
plt.show()

# Bedtime Preference
plt.figure(figsize=(8, 4))
ax2 = sns.countplot(x='Bedtime Preference', data=df, palette=custom_palette)
plt.title('Bedtime Preference', pad=20)
annotate_percentage(ax2)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('figures/bedtime_preference.png')
plt.show()

# WakeUp Time Preference
plt.figure(figsize=(10, 5))
ax3 = sns.countplot(x='WakeUp Time Preference', data=df, palette=custom_palette)
plt.title('WakeUp Time Preference', pad=20)
annotate_percentage(ax3)
ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('figures/wakeup_time_preference.png')
plt.show()

# Planned Study Time
plt.figure(figsize=(10, 5))
ax4 = sns.countplot(x='Planned Study Time', data=df, palette=custom_palette)
plt.title('Planned Study Time', pad=20)
annotate_percentage(ax4)
ax4.spines['top'].set_visible(False)
ax4.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('figures/planned_study_time.png')
plt.show()

# Private Time Requirements
plt.figure(figsize=(8, 4))
ax5 = sns.countplot(x='Private Time Requirements', data=df, palette=custom_palette)
plt.title('Private Time Requirements', pad=20)
annotate_percentage(ax5)
ax5.spines['top'].set_visible(False)
ax5.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('figures/private_time_requirements.png')
plt.show()

# Guest Frequency Preference
plt.figure(figsize=(8, 4))
ax6 = sns.countplot(x='Guest Frequency Preference', data=df, palette=custom_palette)
plt.title('Guest Frequency Preference', pad=20)
annotate_percentage(ax6)
ax6.spines['top'].set_visible(False)
ax6.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('figures/guest_frequency_preference.png')
plt.show()

# Conflict Handling Method
plt.figure(figsize=(10, 6))
ax7 = sns.countplot(x='Conflict Handling Method', data=df, palette=custom_palette)
plt.title('Conflict Handling Method', pad=20)
annotate_percentage(ax7)
ax7.spines['top'].set_visible(False)
ax7.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('figures/conflict_handling_method.png')
plt.show()

# Faculty
plt.figure(figsize=(12, 6))
ax8 = sns.countplot(y='Faculty', data=df, order=df['Faculty'].value_counts().index, palette=custom_palette)
plt.title('Faculty', pad=20)
annotate_percentage(ax8, orientation='horizontal')
ax8.spines['top'].set_visible(False)
ax8.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('figures/faculty.png')
plt.show()

# Attitude towards Roommate Smoking
plt.figure(figsize=(8, 4))
ax9 = sns.countplot(x='Attitude towards Roommate Smoking', data=df, palette=custom_palette)
plt.title('Attitude towards Roommate Smoking', pad=20)
annotate_percentage(ax9)
ax9.spines['top'].set_visible(False)
ax9.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('figures/attitude_roommate_smoking.png')
plt.show()

# Study Time Preference
plt.figure(figsize=(8, 4))
ax10 = sns.countplot(x='Study Time Preference', data=df, palette=custom_palette)
plt.title('Study Time Preference', pad=20)
annotate_percentage(ax10)
ax10.spines['top'].set_visible(False)
ax10.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('figures/study_time_preference.png')
plt.show()

# Communication Preference with Roommate
plt.figure(figsize=(8, 4))
ax11 = sns.countplot(x='Communication Preference with Roommate', data=df, palette=custom_palette)
plt.title('Communication Preference with Roommate', pad=20)
annotate_percentage(ax11)
ax11.spines['top'].set_visible(False)
ax11.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('figures/communication_preference.png')
plt.show()
```

```python
# Function to calculate count, percentage, and top value
def count_and_percentage_with_top(x):
    top_value = x.value_counts().idxmax()  # Most frequent value
    count = x.value_counts().iloc[0]  # Count of the most frequent value
    percentage = (count / len(x)) * 100  # Percentage of the most frequent value
    return f"{top_value}: {count} ({percentage:.2f}%)"

# Group by cluster_label and aggregate statistics
grouped = merged_df.groupby('cluster_label').agg({
    'Bedtime_Preference': count_and_percentage_with_top,
    'Wake_Up_Time_Preference': count_and_percentage_with_top,
    'Planned_Study_Time_Preference': count_and_percentage_with_top,
    'Private_Time_Requirements': count_and_percentage_with_top,
    'Guest_Frequency_Preference': count_and_percentage_with_top,
    'Faculty': count_and_percentage_with_top,
    'Attitude_towards_Roommate_Smoking': count_and_percentage_with_top,
    'Ideal_Study_Environment_Description': count_and_percentage_with_top,
    'Attitude_towards_Borrowing_Sharing': count_and_percentage_with_top,
    'Description_of_Personal_Room_At_Home': count_and_percentage_with_top,
    'Desired_Room_Attributes': count_and_percentage_with_top,
    'Study_Time_Preference': count_and_percentage_with_top,
    'Conflict_Handling_Method': count_and_percentage_with_top,
    'Communication_Preference_with_Roommate': count_and_percentage_with_top,
    'Age': 'mean'  # Mean of age
}).rename(columns={
    'Bedtime_Preference': 'Bedtime Preference (Top Value: Count (Percentage))',
    'Wake_Up_Time_Preference': 'Wake Up Time Preference (Top Value: Count (Percentage))',
    'Planned_Study_Time_Preference': 'Planned Study Time Preference (Top Value: Count (Percentage))',
    'Private_Time_Requirements': 'Private Time Requirements (Top Value: Count (Percentage))',
    'Guest_Frequency_Preference': 'Guest Frequency Preference (Top Value: Count (Percentage))',
    'Faculty': 'Faculty (Top Value: Count (Percentage))',
    'Attitude_towards_Roommate_Smoking': 'Attitude towards Roommate Smoking (Top Value: Count (Percentage))',
    'Ideal_Study_Environment_Description': 'Ideal Study Environment Description (Top Value: Count (Percentage))',
    'Attitude_towards_Borrowing_Sharing': 'Attitude towards Borrowing Sharing (Top Value: Count (Percentage))',
    'Description_of_Personal_Room_At_Home': 'Description of Personal Room At Home (Top Value: Count (Percentage))',
    'Desired_Room_Attributes': 'Desired Room Attributes (Top Value: Count (Percentage))',
    'Study_Time_Preference': 'Study Time Preference (Top Value: Count (Percentage))',
    'Conflict_Handling_Method': 'Conflict Handling Method (Top Value: Count (Percentage))',
    'Communication_Preference_with_Roommate': 'Communication Preference with Roommate (Top Value: Count (Percentage))',
    'Age': 'Average Age'
})

# Display the aggregated statistics
print("Descriptive Analysis by Cluster:")
print(grouped)
```
